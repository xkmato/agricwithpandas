```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
wdi = pd.read_csv('wdi.csv')
```

```python
wdi.shape
```

```python
indicator_map = wdi.filter(['Indicator Name', 'Indicator Code'], axis=1)
```

```python
indicator_map = indicator_map.drop_duplicates()
```

```python
indicator_map
```

```python
wdi = wdi.melt(
    id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
    var_name='Year',
    value_name='Indicator Value'
)
```

```python
wdi.head()
```

```python
wdi.tail()
```

```python
wdi.sample(10)
```

```python
wdi.shape
```

```python
wdi.drop(['Country Code', 'Indicator Name'], axis=1, inplace=True)
```

```python
wdi.head()
```

```python
df = wdi.pivot_table('Indicator Value', ['Country Name', 'Year'], 'Indicator Code')
```

```python
df.head(22)
```

```python
df.reset_index(inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
indicator_map
```

# Plot Employment in Agriculture over the years in Uganda

```python
ug_df = df[df['Country Name']=='Uganda']
```

```python
ug_df.plot(x='Year', y='SL.AGR.EMPL.ZS')
```

## Compare men to women

```python
ug_df.plot(x='Year', y=['SL.AGR.EMPL.MA.ZS', 'SL.AGR.EMPL.FE.ZS'], kind='bar', figsize=(20, 10))
```

# Compare to Kenya

```python
ke_df = df[df['Country Name']=='Kenya']
```

```python
ke_df
```

```python
fig, ax = plt.subplots()
ug_df.plot(ax=ax, x='Year', y='SL.AGR.EMPL.ZS', label='Uganda')
ke_df.plot(ax=ax, x='Year', y='SL.AGR.EMPL.ZS', label='Kenya')
```

# Track Average rate of deforestration in Uganda between 1998-2016 in sq. km/year

```python
forest_1998 = ug_df[ug_df['Year']=='1998']['AG.LND.FRST.K2'].iloc[0]
forest_2016 = ug_df[ug_df['Year']=='2016']['AG.LND.FRST.K2'].iloc[0]
forests_cut_down = forest_1998 - forest_2016
forests_cut_down, forest_1998, forest_2016
```

# How much forest has been cleared for non-agriculture purpose

```python
ug_df['Arable Land sq. km'] = ug_df['AG.LND.ARBL.HA']*0.01
```

```python
arable_1998 = ug_df[ug_df['Year']=='1998']['Arable Land sq. km'].iloc[0]
arable_2016 = ug_df[ug_df['Year']=='2016']['Arable Land sq. km'].iloc[0]
arable_land_created = arable_2016 - arable_1998
arable_land_created, arable_2016, arable_1998
```

```python
forests_cut_down - arable_land_created
```
