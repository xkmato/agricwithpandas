````python
import pandas as pd
import matplotlib.pyplot as plt
````

# Introducing the data

We first take a look at our data to get a first impression. Pandas provides 3 convinent method that allow you to peep at the data
* `head()` - Allows you to see the first rows of your dataframe
* `tail()` - Allows you to see the last rows of your dataframe
* `sample(n)` - Allows you to see a random sample of `n` rows of your dataframe

````python
wdi = pd.read_csv('wdi.csv')
````

````python
wdi.head(10)
````

````python
wdi.tail()
````

````python
wdi.sample(5)
````

## You can create a new dataframe from a subset of your current dataframe

We now want to store indicators and their names in a sepate dataframe. We want to be able to refer to this in the future

````python
indicators = wdi[['Indicator Name', 'Indicator Code']]
````

## Droping duplicate values

Using `drop_duplicates()`, we can remove any duplicate rows from our dataframe.


````python
indicators.drop_duplicates(inplace=True)
````

````python
indicators
````

## Unpivoting the data

Unpivoting means turning a couple of column names into values of a specific row. For example in the scenario, we want to create a new column `Year` and give it values of the column names that are years.

We shall use the `melt()` see documentation [here](https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.melt.html)

````python
wdi1 = wdi.melt(
    id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
    var_name='Year',
    value_name='Indicator Value'
)
````

````python
wdi1.head()
````

## Droping columns

Now that we have our `indicators dataframe`, we no longer need both the `Indicator Name` and `Indicator Code` columns since they have the same information. We also do not need both the `Country Code` and `Country Name` columns for the same reason.

We shall now use the `drop()` to remove `Country Code` and `Indicator Name` columns

````python
wdi1.drop(['Country Code', 'Indicator Name'], axis=1, inplace=True)
````

````python
wdi1.head(10)
````

## Pivoting the table

What if we wanted to make some of the values columns? The opposite of what we did earlier with `melt`. This is called pivoting.

What we want to do is make all `Indicator Codes` columns on our dataframe

We shall use the `pivot_table()`. See documentation [here](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html)

````python
df = wdi1.pivot_table('Indicator Value', ['Country Name', 'Year'], 'Indicator Code')
````

````python
df.reset_index(inplace=True)
df.head()
````

## Understanding our data's metdata

Now that our data looks just the way we like it to look, now we can look at the metadata to get more indepth information on the general summary of the data

We shall use `describe()` method to see a stastical summary of the data. We shall also user `info()` method to see the datatypes of each column

````python
df.describe()
````

````python
df.info()
````

## Modifying a column

We would like to modify our `Year` column from type `object` to type integer. Since years are always numbers

````python
df['Year'] = pd.to_numeric(df['Year'])
````

````python
df.info()
````

# 1. Plot the % of Ugandan's employed in Agriculture

Demostration of a simple plot using `pandas.plot()`

````python
ug_df = df[df['Country Name']=='Uganda']
````

````python
ug_df.head()
````

````python
ug_df.plot(x='Year', y='SL.AGR.EMPL.ZS', figsize=(10,8), kind='bar')
````

## Compare % of women vs men employed in Agriculture

Now we plot a graph with multiple variables on the Y axis. It is a simple as passing a list of column names in to the `y` kwarg in `pandas.plot()`

````python
ug_df.plot(x='Year', y=['SL.AGR.EMPL.MA.ZS', 'SL.AGR.EMPL.FE.ZS'], kind='bar', figsize=(10, 8))
````

## Compare Kenya to Uganda

We plot multiple graphs on the same figure.

````python
ke_df = df[df['Country Name']=='Kenya']
````

````python
fig, ax = plt.subplots()
ug_df.plot(ax=ax, x='Year', y='SL.AGR.EMPL.ZS', label='Uganda',figsize=(15,8))
ke_df.plot(ax=ax, x='Year', y='SL.AGR.EMPL.ZS', label='Kenya', figsize=(15,8))
````

# 2. What is the average deforestration rate in sq. km/year in Uganda between 1998-2016?

How do we access datapoints in the dataframe and how do we do simple calculations with them?

````python
forest_1998 = ug_df[ug_df['Year']==1998]['AG.LND.FRST.K2'].iloc[0]
forest_2016 = ug_df[ug_df['Year']==2016]['AG.LND.FRST.K2'].iloc[0]
forests_cut_down = forest_1998 - forest_2016
forests_cut_down, forest_1998, forest_2016
````

````python
number_of_years = 19
average_deforestation_per_year = forests_cut_down/number_of_years
average_deforestation_per_year
````

## How much forest has been cleared for a reason other than agriculture?

How to create a new column derived from an already existing column on the dataframe

````python
ug_df['Arable Land sq. km'] = ug_df['AG.LND.ARBL.HA']*0.01
````

````python
arable_1998 = ug_df[ug_df['Year']==1998]['Arable Land sq. km'].iloc[0]
arable_2016 = ug_df[ug_df['Year']==2016]['Arable Land sq. km'].iloc[0]
arable_land_created = arable_2016 - arable_1998
arable_land_created, arable_2016, arable_1998
````

````python
forests_cut_down - arable_land_created
````

## Compare Uganda deforestation to the Gambia deforestation

````python
ga_df = df[df['Country Name'] == 'Gambia, The']
````

````python
fig1, ax1 = plt.subplots()
ug_df.plot(ax=ax1, x='Year', y='AG.LND.FRST.K2', label='Uganda', figsize=(15,8))
ga_df.plot(ax=ax1, x='Year', y='AG.LND.FRST.K2', label='Gambia', figsize=(15,8))
````

# 3. Plot the productivity of France's land under cereal production over the years

Some times a graph does not share any useful information until you compare it to another graph.

How much cereal does France produce on each piece of land dedicated to cereal production?

````python
fr_df = df[df['Country Name']=='France']
fr_df['Cereal/Hectare'] = fr_df['AG.PRD.CREL.MT']/fr_df['AG.LND.CREL.HA']
fr_df.plot(x='Year', y='Cereal/Hectare', figsize=(15,8))
````

## Compare that to Uganda

Does Uganda produce the same amount of cereal per unit piece of land as France?

````python
ug_df['Cereal/Hectare'] = ug_df['AG.PRD.CREL.MT']/ug_df['AG.LND.CREL.HA']

fig2, ax2 = plt.subplots()
ug_df.plot(ax=ax2, x='Year', y='Cereal/Hectare', label='Uganda', figsize=(15,8))
fr_df.plot(ax=ax2, x='Year', y='Cereal/Hectare', label='France', figsize=(15,8))
````
